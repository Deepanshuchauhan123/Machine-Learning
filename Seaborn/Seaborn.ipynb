# Seaborn Python
## Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# importing seaborn
import seaborn as sns

# setting seaborn style
sns.set_style("whitegrid")

#data
data = pd.read_csv("sales_data_sample.csv")
```

```python
data.head()
```

```python
# to plot the Distribution plot with histogram
sns.distplot(data['MSRP'])
plt.show()
```

```python
# Distribution plot without histogram
sns.distplot(data['MSRP'],hist=False)
plt.show()
```

```python
#subplot 1
plt.subplot(2,2,1)
sns.distplot(data['SALES'])

#subplot 2
plt.subplot(2,2,2)
sns.distplot(data['MSRP'])

#subplot 3
plt.subplot(2,2,3)
sns.distplot(data['PRICEEACH'])

#subplot 4
plt.subplot(2,2,4)
sns.distplot(data['QUANTITYORDERED'])

plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10,8))
sns.boxplot(data['QUANTITYORDERED'])
plt.show()
```

## JointPlot

```python
sns.jointplot('QUANTITYORDERED','SALES',data)
plt.show()
```

```python
# Decreasing the range of data values for better view 
data1 = data[(data.QUANTITYORDERED<60)& (data.SALES<10000)]
```

```python
sns.jointplot('QUANTITYORDERED','SALES',data1)
plt.show()
```

```python
dataframe = pd.read_csv("1000 Sales Records.csv")
dataframe.head()
```

```python
plt.figure(figsize=(18,8))
sns.boxplot(x='Item Type',y='Total Profit',data=dataframe)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='Item Type',y='Unit Price',data=dataframe)
plt.title('Average Price')
plt.show()
plt.figure(figsize=(16,8))
sns.barplot(x='Item Type',y='Unit Price',data=dataframe,estimator=np.median)
plt.title('Median Profit')
plt.show()
```
