```python
import numpy as np
import pandas as pd

df=pd.read_csv("Melbourne.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
```

```python
# checking whether data is null or not
df.isnull()
```

## Identifying missing values in Columns
### you can identify total missing values in a columns by using df.isnull().sum() method

```python
# for each column check how many missing values are there
df.isnull().sum()
```

## Identifying missing values in rows

```python
# is there any missing value in a row
df.isnull().any(axis=1)
```

```python
# rows having all the missing values
df.isnull().all(axis=1)
```

```python
# sum of missing values in row
df.isnull().sum(axis=1)
```

## Treating missing values in a column

## various methods for missing values are:-
### 1. Deletions* (Best prefer)
### 2.Imputation ---
    2.1 Mean
    2.2 Mode
    2.3 Median

```python
# summing up the missing values (column-wise)
# (missing values/total values)/100 and taking 2 digits after point 
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
# removing 3 columns
df=df.drop('BuildingArea',axis=1)
df=df.drop('YearBuilt',axis=1)
df=df.drop('CouncilArea',axis=1)
round(100*(df.isnull().sum()/len(df.index)),2)
```
