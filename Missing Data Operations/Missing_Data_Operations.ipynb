```python
import numpy as np
import pandas as pd

df=pd.read_csv("Melbourne.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
```

```python
# checking whether data is null or not
df.isnull()
```

## Identifying missing values in Columns
### you can identify total missing values in a columns by using df.isnull().sum() method

```python
# for each column check how many missing values are there
df.isnull().sum()
```

## Identifying missing values in rows

```python
# is there any missing value in a row
df.isnull().any(axis=1)
```

```python
# rows having all the missing values
df.isnull().all(axis=1)
```

```python
# sum of missing values in row
df.isnull().sum(axis=1)
```

## Treating missing values in a column

## various methods for missing values are:-
### 1. Deletions* (Best prefer)
### 2.Imputation ---
    2.1 Mean
    2.2 Mode
    2.3 Median

```python
# summing up the missing values (column-wise)
# (missing values/total values)/100 and taking 2 digits after point 
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
# removing 3 columns
df=df.drop('BuildingArea',axis=1)
df=df.drop('YearBuilt',axis=1)
df=df.drop('CouncilArea',axis=1)
round(100*(df.isnull().sum()/len(df.index)),2)
```

## Treating missing values in row

```python
# rows having missing values more than 5
df[df.isnull().sum(axis=1)>5]
```

```python
# total rows having missing data in more than 5 columns
len(df[df.isnull().sum(axis=1)>5].index)
```

```python
# percentage of missing data in row having more than 5 Nan values
(len(df[df.isnull().sum(axis=1)>5].index)/len(df.index))*100
```

Thus we can remove these 18% data as they are having more missing values

```python
# retaining the rows having atmost 5 missing values
df=df[df.isnull().sum(axis=1)<=5]

# looking at the summary again
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
# removing price nan values
df=df[~np.isnan(df['Price'])]

# looking at the summary again
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
# the amount of data we are left with
len(df.index)
```

```python
df['Landsize'].describe()
```

As it is having huge difference between min and max and 25,50,75% and max. SO if we impute then it will be a huge problem.
So a better approach will be removing the corresponding rows from the Landsize.

```python
# droping the missing row of Lansize
df=df[~np.isnan(df['Landsize'])]

# looking at the summary again
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
# left data
len(df.index)
```

```python
df[np.isnan(df['Lattitude'])]
```

```python
df.loc[:,['Lattitude','Longtitude']].describe()
```

```python
# imputing Lattitude and Long. values by mean vales
df.loc[np.isnan(df['Lattitude']),['Lattitude']] = df['Lattitude'].mean()
df.loc[np.isnan(df['Longtitude']),['Longtitude']] = df['Longtitude'].mean()

# looking at the summary again
round(100*(df.isnull().sum()/len(df.index)),2)
```
