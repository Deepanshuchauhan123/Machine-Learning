# Matplotlib
### Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays
###  It allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5,110,100)
```

```python
y = np.linspace(10,1000,100)
```

```python
plt.plot(x,y)
```

## Graph Labelling

```python
x = [2,4,5,6,8,10]
y = [6,3,2,7,9,5]
```

```python
plt.plot(x,y)
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
plt.title("Random Graph")
```

## Graph lines and color

```python
x = np.linspace(5,20,10)
y = np.linspace(8,100,10)
```

```python
# first character is color i.e 'r' means RED and shape is '+'
plt.plot(x,y,'r+')
```

## Multiple Graphs in a single graph

```python
x = np.linspace(5,20,10)
y = np.linspace(2,3,10)
```

```python
# we can show n-graphs in a single graph use comma seperation
plt.plot(x,y,'b+', x,y**2,'g^', x,y**3,'c*')
```

## Graph side by side

```python
x = np.linspace(1,10,100)
y = np.log(x)
```

```python
# initiate a new figure explicitly
plt.figure(1)

# subplot refers to create a sub plot with 1 row and 2 columns

# create the first subplot in figure 1
# equivalent to (1,2,1)
plt.subplot(121)
plt.title("Y = Log(X)")
plt.plot(x,y)

# create the second plot 1 rows 2 columns position 2nd
plt.subplot(122)
plt.title("Y=Log(X)**2")
plt.plot(x,y**2)

plt.show()
```

```python
x = np.linspace(1,10,100)

# optional command, since matplotlib creates a figure by default anyway
plt.figure(1)

#subplot 1

plt.subplot(121)
plt.title("Linear")
plt.plot(x,x)

# subplot 2
plt.subplot(122)
plt.title("Cubic")
plt.plot(x,x**3)

# To change to the next row 
plt.figure(2)

# subplot 3
plt.subplot(121)
plt.title("Log(X)")
plt.plot(x,np.log(x))

# subplot 4

plt.subplot(122)
plt.title("Exponential")
plt.plot(x,x**2)
```

```python
data = pd.read_csv("sales_data_sample.csv")
data.head()
```

## Notes
#### * Quartile is the portion of the data        

#### * 0------25(Q1)-------50(Q2)---------75(Q3)-----------100(Q4)                                                                                            
#### * Outliers are data points that are far from other data points. In other words, they're unusual values in a dataset.
